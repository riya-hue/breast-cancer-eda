```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("data.csv")
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True)

# Encode target for easier plotting
df['diagnosis'] = df['diagnosis'].map({'M':'Malignant','B':'Benign'})
```

```python
# 1. Countplot: Number of Malignant vs Benign
sns.countplot(x='diagnosis', data=df, palette='Set2')
plt.title("Number of Malignant vs Benign Tumors")
plt.show()
```

```python
# 2. Histogram: Distribution of radius_mean
plt.hist(df['radius_mean'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Radius Mean")
plt.xlabel("Radius Mean")
plt.ylabel("Count")
plt.show()
```

```python
#3. Boxplot: Compare area_mean by diagnosis
sns.boxplot(x='diagnosis', y='area_mean', data=df, palette='Set1')
plt.title("Area Mean by Diagnosis")
plt.show()
```

```python
# 4. Scatter plot: radius_mean vs texture_mean
plt.scatter(df['radius_mean'], df['texture_mean'], c=df['diagnosis'].map({'Malignant':1,'Benign':0}), cmap='coolwarm', alpha=0.6)
plt.title("Radius vs Texture by Diagnosis")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.show()
```

```python
# 5. Pairplot: Sample of selected features
sns.pairplot(df[['radius_mean','texture_mean','area_mean','smoothness_mean','diagnosis']], hue='diagnosis', palette='Set2')
plt.show()
```

```python
# 7. Pie chart: Proportion of Tumor Types
df['diagnosis'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title("Proportion of Malignant vs Benign Tumors")
plt.ylabel("")
plt.show()
```

```python
# 8. Violin Plot: Smoothness by Diagnosis
sns.violinplot(x='diagnosis', y='smoothness_mean', data=df, palette='Set3')
plt.title("Smoothness Mean by Tumor Type")
plt.show()
```
